# 1. Mengambil data dari ucimlrepo

```python
import pandas as pd
from ucimlrepo import fetch_ucirepo

pisweb = fetch_ucirepo(id=327)
X = pisweb.data.features
y = pisweb.data.targets
df = pd.concat([X, y], axis=1)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['result'].value_counts()
```

# 2. periksa distribusi kelas target

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(6, 4))
sns.countplot(x='result', data=df, palette='viridis')
plt.title('Distribusi Kelas Target')
plt.xlabel('Kelas')
plt.ylabel('Jumlah')
plt.show()
```

# 3. Deteksi oulier dengan menggunakan Z-Score

```python
import numpy as np
from scipy import stats

z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))
outliers = (z_scores > 3).any(axis=1)
print("\nJumlah Outlier:", outliers.sum())
```

```python
df_cleaned = df[~outliers]
```

# 4. Split fitur dan target setelah cleaing

```python
X_cleaned = df_cleaned.drop(columns=['result'])
y_cleaned = df_cleaned['result']
```

# 5. pembuatan model dan training
  - split data menjadi train dan test
  - Kemudian latih model dengaan RandomForestClassifier

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y_cleaned, test_size=0.2, random_state=42)
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python

y_pred = model.predict(X_test)
```

# 6. evaluasi model

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print(f"\nAkurasi Model: {accuracy_score(y_test, y_pred) * 100:.2f}%")
```

# 7. Visualiasi pitur penting

```python
feature_importances = model.feature_importances_
feature_names = X_cleaned.columns
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df, palette='viridis')
plt.title('Fitur Penting dalam Model')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
